```python
import pandas as pd
import numpy as np
```

```python
import glob

path = './data' # sample csv files in folder data

allfiles = glob.glob(path+ "/*.csv")
weibo = pd.DataFrame()
list_ = []
for file_ in allfiles:
    df = pd.read_csv(file_,index_col=None, header=0)
    list_.append(df)
weibo = pd.concat(list_)

cellphone =  weibo.dropna(axis=0, how='any', subset=['term']) 
cellphone =  weibo.dropna(axis=0, how='any', subset=['sex']) 
cellphone  = cellphone.reset_index()
cellphone['term'].replace(' ', np.nan, inplace=True)
cellphone.dropna(subset=['term'], inplace=True)

redundant_list = ['粉丝红包', '生日动态', '用户任务', '微博等级', '熊猫守护者', \
                 '分享按钮', '微博会员中心', '微话题', '微博故事', '微博活动',\
                 '红包活动','Weibo.intl','微博 weibo.com', '微博 HTML5 版',\
                 '微博问答','微博体育', '网易云音乐', '360安全浏览器', '内涵段子APP',\
                 '微博品牌活动', '今日头条', '阴阳师', '微盘', '微博电影','微博客户端',\
                 '无线校园活动','支付宝', '荒野行动', '微博红包', '哔哩哔哩移动客户端',\
                 '秒拍客户端', '搜狗高速浏览器', '微博音乐', '全民K歌','微博极速版', '哔哩哔哩', \
                  '新浪博客', '微博运动', '美拍', '专业版微博', '能量榜', '微卡券','QQ音乐',\
                  '快手','唱吧','新浪新闻','粉丝头条', '我是答题王', '抖音短视频', '百度',\
                 '派派','陌陌', '腾讯新闻', '优酷土豆', 'Keep', '凤凰新闻', '优酷', '知乎客户端',\
                 '豆瓣', '手机淘宝', '微博手机版', '手机微博触屏版', 'UC浏览器Android', '炫舞手游', \
                  '微博钱包福利', '微博股票', 'UC浏览器电脑版', '即刻', '小红书','红人淘', '秒拍网页版',\
                 '终结者2','手机NBA', '明星势力榜', 'LOFTER', '网易', '土豆', 'AcFun', 'miyijia测试君',\
                 '可可英语', '券白白', 'Weico.Android', '一直播', '美图', '京东', '淘宝', '新浪游戏',\
                 '天猫', '单词', '微博搜索', '酷狗', '爱奇艺', '新闻','美团','搜狐','腾讯', '有信']

for redundant in redundant_list:
    cellphone = cellphone[cellphone["term"].str.contains(redundant) == False]
    
    
iphone_list = ['iphone', 'iPhone']
ipad_list = ['iPad', 'ipad']
huawei_list = ['华为', 'Huawei', 'HUAWEI','荣耀', '麦芒']
oppo_list = ['OPPO'] #欧珀
vivo_list = ['vivo'] #步步高
xiaomi_list = ['小米','红米']
samsung_list = ['三星', 'Samsung']
meizu_list = ['魅族', '魅蓝']
oneplus_list = ['OnePlus'] #一加
leshi_list = ['乐1s','乐2', '乐Max 2']
threesixzero_list = ['360手机']
smartisan_list = ['坚果 Pro','Smartisan']
gionee_list = ['金立智能手机', '金立']
coolpad_list = ['酷派手机', 'dazen', 'Dazen', '大神', '酷派', 'Cool', 'cool1']
zte_list = ["努比亚", 'nubia', 'ZTE', '中兴']
nokia_list = ['Nokia', '诺基亚']
lenovo_list = ['联想智能手机','联想ZUK']
htc_list = ['HTC']
sony_list = ['Xperia', 'XPERIA']

for phone in iphone_list:
    cellphone.loc[cellphone['term'].str.contains(phone),'term'] = 'IPHONE'
    
for phone in huawei_list:
    cellphone.loc[cellphone['term'].str.contains(phone),'term'] = 'HUAWEI'
    
for phone in oppo_list:
    cellphone.loc[cellphone['term'].str.contains(phone),'term'] = 'OPPO'

for phone in vivo_list:
    cellphone.loc[cellphone['term'].str.contains(phone),'term'] = 'VIVO'
    
for phone in xiaomi_list:
    cellphone.loc[cellphone['term'].str.contains(phone),'term'] = 'XIAOMI'

for phone in samsung_list:
    cellphone.loc[cellphone['term'].str.contains(phone),'term'] = 'SAMSUNG'
    
for phone in meizu_list:
    cellphone.loc[cellphone['term'].str.contains(phone),'term'] = 'MEIZU'
    
for phone in oneplus_list:
    cellphone.loc[cellphone['term'].str.contains(phone),'term'] = 'ONEPLUS'
    
for phone in leshi_list:
    cellphone.loc[cellphone['term'].str.contains(phone),'term'] = 'LESHI'
    
for phone in threesixzero_list:
    cellphone.loc[cellphone['term'].str.contains(phone),'term'] = 'THREESIXZERO'
    
for phone in smartisan_list:
    cellphone.loc[cellphone['term'].str.contains(phone),'term'] = 'SMARTISAN'
    
for phone in gionee_list:
    cellphone.loc[cellphone['term'].str.contains(phone),'term'] = 'GIONEE'
    
for phone in coolpad_list:
    cellphone.loc[cellphone['term'].str.contains(phone),'term'] = 'COOLPAD'

for phone in zte_list:
    cellphone.loc[cellphone['term'].str.contains(phone),'term'] = 'ZTE'
    
for phone in ipad_list:
    cellphone.loc[cellphone['term'].str.contains(phone),'term'] = 'IPAD'
    
for phone in nokia_list:
    cellphone.loc[cellphone['term'].str.contains(phone),'term'] = 'NOKIA'
        
for phone in lenovo_list:
    cellphone.loc[cellphone['term'].str.contains(phone),'term'] = 'LENOVO'

for phone in sony_list:
    cellphone.loc[cellphone['term'].str.contains(phone),'term'] = 'SONY'

for phone in htc_list:
    cellphone.loc[cellphone['term'].str.contains(phone),'term'] = 'HTC'
    
phone_list = ['IPHONE', 'HUAWEI', 'OPPO', 'VIVO', 'XIAOMI', 'SAMSUNG', \
              'GIONEE', 'MEIZU', 'ONEPLUS', 'ZTE', 'COOLPAD','THREESIXZERO',\
             'NOKIA', 'LENOVO', 'SONY', 'HTC','SMARTISAN','LESHI']

cellphone = cellphone[cellphone['term'].str.contains('|'.join(phone_list))]
cellphone = cellphone.reset_index()
cellphone_clean = cellphone.drop(['level_0', 'index'], axis=1)
```

```python
china_cities = pd.read_table("china_city_loc.txt", sep=' ', header=None, names = ["city", "longitude", "latitude"])
location_list = cellphone_clean["loc"].values
location_dict = list(set(location_list))
locations = []
for i in location_dict:
    j1 = i.replace(' ','')
    j2 = j1.replace('市','')
    j = j2.replace('区','')
    locations.append(j)
    
longitudes = []
latitudes = []

for i, loc in enumerate(locations):
    lng = 0.0
    lat = 0.0
    match = False
    for j in range(china_cities.shape[0]):
        if loc == china_cities["city"][j]:
            lng = china_cities["longitude"][j]
            lat = china_cities["latitude"][j]
            match = True
            break
          
    longitudes.append(lng)
    latitudes.append(lat)

longitudes = list(map(float, longitudes))
latitudes = list(map(float, latitudes))
   
lng_dict = dict(zip(location_dict, longitudes))
lat_dict = dict(zip(location_dict, latitudes))

lng_list = []
lat_list = []
for i in range(cellphone_clean.shape[0]):
    loc = cellphone_clean["loc"][i]
    lng = lng_dict[loc]
    lat = lat_dict[loc]
    lng_list.append(lng)
    lat_list.append(lat)

cellphone_clean["lng"] = lng_list
cellphone_clean["lat"] = lat_list
```

```python
# http://gmt-china.org/datas/
CNborder = np.loadtxt("./CN-border-L1.dat")
CNborder.shape
```

```python
huawei_user = cellphone_clean[cellphone_clean["term"].str.contains('HUAWEI') == True]
iphone_user = cellphone_clean[cellphone_clean["term"].str.contains('IPHONE') == True]
oppo_user = cellphone_clean[cellphone_clean["term"].str.contains('OPPO') == True]
vivo_user = cellphone_clean[cellphone_clean["term"].str.contains('VIVO') == True]
xiaomi_user = cellphone_clean[cellphone_clean["term"].str.contains('XIAOMI') == True]
```

```python
iphone_stat = [iphone_user['weibo_num'].mean(), iphone_user['following'].mean(), iphone_user['follower'].mean()]
huawei_stat = [huawei_user['weibo_num'].mean(), huawei_user['following'].mean(), huawei_user['follower'].mean()]
oppo_stat = [oppo_user['weibo_num'].mean(), oppo_user['following'].mean(), oppo_user['follower'].mean()]
vivo_stat = [vivo_user['weibo_num'].mean(), vivo_user['following'].mean(), vivo_user['follower'].mean()]
xiaomi_stat = [xiaomi_user['weibo_num'].mean(), xiaomi_user['following'].mean(), xiaomi_user['follower'].mean()]
```

```python
import matplotlib.pyplot as plt
import math

lat_min =  18
lat_max = 55
lng_min = 75
lng_max = 135
lat_avg = (lat_min + lat_max) / 2

xmin = lng_min
xmax = lng_max
ymin = lat_min
ymax = lat_max

dx = xmax - xmin
dy = ymax - ymin

lon_scale = math.cos( lat_avg /180 * math.pi)


plt.figure(figsize=(dx*lon_scale*0.3, dy*0.3))

plt.xlim(xmin, xmax)
plt.ylim(ymin, ymax)

plt.title('Top Five Cell Phone Brands in China as Seen from Weibo Users \n One Dot per User, April 2018',fontsize=25)
plt.xlabel("longitude",fontsize=20)
plt.ylabel("latitude",fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

#noise = 0.2
#dlng = np.random.uniform(-noise, noise, size = len(lng_list))
#dlat = np.random.uniform(-noise, noise, size = len(lat_list))
#plt.plot(lng_list+dlng,lat_list+dlat, 'r.')

noise = 0.3
dlng = np.random.uniform(-noise, noise, size = iphone_user.shape[0])
dlat = np.random.uniform(-noise, noise, size = iphone_user.shape[0])
plt.plot(iphone_user["lng"].values+dlng,iphone_user["lat"].values+dlat, 'b.',alpha = 1)

noise = 0.3
dlng = np.random.uniform(-noise, noise, size = huawei_user.shape[0])
dlat = np.random.uniform(-noise, noise, size = huawei_user.shape[0])
plt.plot(huawei_user["lng"].values+dlng,huawei_user["lat"].values+dlat, 'r.',alpha = 1)


noise = 0.3
dlng = np.random.uniform(-noise, noise, size = oppo_user.shape[0])
dlat = np.random.uniform(-noise, noise, size = oppo_user.shape[0])
plt.plot(oppo_user["lng"].values+dlng,oppo_user["lat"].values+dlat, 'g.',alpha = 1)

noise = 0.3
dlng = np.random.uniform(-noise, noise, size = vivo_user.shape[0])
dlat = np.random.uniform(-noise, noise, size = vivo_user.shape[0])
plt.plot(vivo_user["lng"].values+dlng,vivo_user["lat"].values+dlat, 'yx',alpha = 1)

noise = 0.3
dlng = np.random.uniform(-noise, noise, size = xiaomi_user.shape[0])
dlat = np.random.uniform(-noise, noise, size = xiaomi_user.shape[0])
plt.plot(xiaomi_user["lng"].values+dlng,xiaomi_user["lat"].values+dlat, 'm+',alpha = 1)



plt.scatter(CNborder[:,0],CNborder[:,1], c='k', marker='.', s=0.05)

plt.tight_layout()
plt.legend(['iPhone', 'huawei', 'oppo', 'vivo', 'xiaomi'],frameon=False, fontsize=20, loc="upper center")


plt.savefig("CN_cellphone_users.png",dpi = 500)
#plt.savefig("CN_cellphone_users.pdf")
#plt.savefig("CN_cellphone_users.eps")
plt.show()
```
